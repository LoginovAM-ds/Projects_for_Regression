```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
data=pd.read_csv('AAPL.csv')
data.tail(1)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
col_for_candle = {'Open Price': 'Open',
                  'High Price': 'High', 'Low Price': 'Low', 'Close Price': 'Close'}
for_candle=data.rename(columns=col_for_candle)
```

```python

plt.figure(figsize=(14,8))
plt.plot(data['Date'], data['Close Price'])
plt.xlabel('Date')
plt.grid(True)
```

```python
sma30=pd.DataFrame()
sma30['Adj Close Price'] = data['Adj Close Price'].rolling(window=30).mean()
sma100 = pd.DataFrame()
sma100['Adj Close Price']=data['Adj Close Price'].rolling(window=100).mean()
```

```python
plt.figure(figsize=(14,8))
plt.plot(data['Adj Close Price'],label='data')
plt.plot(sma30['Adj Close Price'], label='30')
plt.plot(sma100['Adj Close Price'], label='100')
plt.legend(loc='upper left')
```
